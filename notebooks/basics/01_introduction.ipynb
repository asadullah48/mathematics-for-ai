## Linear Algebra Basics {#linear-algebra}

Linear algebra deals with vectors, matrices, and linear transformations.

```python
# Vector operations
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

# Dot product
dot_product = LinearAlgebra.dot_product(v1, v2)
print(f"Dot product: {dot_product}")

# Norms
l1_norm = LinearAlgebra.norm(v1, ord=1)
l2_norm = LinearAlgebra.norm(v1, ord=2)
print(f"L1 norm: {l1_norm}, L2 norm: {l2_norm}")

# Cosine similarity
similarity = LinearAlgebra.cosine_similarity(v1, v2)
print(f"Cosine similarity: {similarity:.4f}")
```

```python
# Matrix operations
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# Matrix multiplication
C = LinearAlgebra.matrix_multiply(A, B)
print(f"Matrix product:\n{C}")

# Determinant
det = LinearAlgebra.determinant(A)
print(f"Determinant of A: {det}")

# Eigen decomposition
eigenvalues, eigenvectors = LinearAlgebra.eigendecomposition(A)
print(f"Eigenvalues: {eigenvalues}")
print(f"Eigenvectors:\n{eigenvectors}")
```

## Calculus Fundamentals {#calculus}

Calculus is essential for understanding optimization and learning algorithms.

```python
# Numerical differentiation
f = lambda x: x ** 2
derivative = Calculus.numerical_derivative(f, 3.0)
print(f"Derivative of x² at x=3: {derivative:.4f} (exact: 6.0)")

# Gradient of multivariate function
g = lambda x: x[0] ** 2 + x[1] ** 2
gradient = Calculus.numerical_gradient(g, np.array([3.0, 4.0]))
print(f"Gradient of x²+y² at (3,4): {gradient}")
```

```python
# Numerical integration
f = lambda x: np.sin(x)
integral = Calculus.simpson(f, 0, np.pi, n=100)
print(f"Integral of sin(x) from 0 to π: {integral:.6f} (exact: 2.0)")

# Root finding
f = lambda x: x ** 2 - 2
root = Calculus.newton_raphson(f, 1.5)
print(f"Root of x²-2: {root:.10f} (exact: {np.sqrt(2):.10f})")
```

## Probability & Statistics {#probability}

Probability theory is crucial for understanding uncertainty and making predictions.

```python
# Probability distributions
x = np.linspace(-3, 3, 100)

# Gaussian PDF
pdf = Probability.gaussian_pdf(x, mu=0, sigma=1)

# Plot
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x, pdf, 'b-', linewidth=2, label='N(0,1)')
ax.set_title('Gaussian Distribution')
ax.set_xlabel('x')
ax.set_ylabel('PDF')
ax.legend()
ax.grid(True, alpha=0.3)
plt.show()
```

```python
# Statistics
data = np.random.randn(1000) * 2 + 5  # Normal data with mean=5, std=2

# Summary statistics
summary = Statistics.summary(data)
print(f"Count: {summary.count}")
print(f"Mean: {summary.mean:.4f}")
print(f"Std: {summary.std:.4f}")
print(f"Median: {summary.median:.4f}")
print(f"Skewness: {summary.skewness:.4f}")
```

## Getting Started with ML Models {#getting-started}

```python
# Linear Regression example
np.random.seed(42)
X = np.random.randn(100, 2)
y = 2 * X[:, 0] + 3 * X[:, 1] + np.random.randn(100) * 0.1

model = LinearRegression(learning_rate=0.1, n_iterations=1000)
model.fit(X, y)

print(f"Learned weights: {model.weights}")
print(f"True weights: [2.0, 3.0]")
print(f"R² score: {model.score(X, y):.4f}")
```

```python
# Neural Network example
from sklearn.datasets import make_classification

# Generate binary classification data
X, y = make_classification(n_samples=500, n_features=10, n_classes=2, random_state=42)

# Create and train neural network
nn = NeuralNetwork(
    layer_sizes=[10, 32, 16, 2],
    activations=['relu', 'relu', 'softmax'],
    learning_rate=0.001,
    optimizer='adam'
)

nn.fit(X, y, epochs=100, batch_size=32, verbose=False)

print(f"Training accuracy: {nn.score(X, y):.4f}")
```
